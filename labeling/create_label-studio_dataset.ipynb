```python
import pandas as pd
import numpy as np
```

```python
# load dataset
dataset1_path = 'labeled_dataset_4688/bc_ab_wildfires.json'
dataset1 = pd.read_json(dataset1_path)
dataset1 = dataset1[['tweet_id', 'img_id', 'posted_at', 'username', 'author_location', 'country_code', 'country', 'full_name', 'text','image', 'label', 'contains_personal_info']]
dataset1['image'] = dataset1['image'].apply(lambda x: x.split('\\')[7])
base_path1 = '/data/local-files/?d=data/bc_ab_images/' #'labeled_dataset_4688/bc_ab_images/'
dataset1['image'] = dataset1['image'].apply(lambda x: base_path1 + x)
# load dataset
dataset2_path = 'labeled_dataset_4688/bc_ab_jasper_07-18_07-25_labeled.csv'
dataset2 = pd.read_csv(dataset2_path, dtype={'media_keys': str})
dataset2 = dataset2.rename(columns={'author_loc': 'author_location'})
dataset2 = dataset2.rename(columns={'author_usrname': 'username'})
dataset2['image'] = dataset2['image'].apply(lambda x: x.split('\\')[7])
base_path2 = '/data/local-files/?d=data/new_images/' #'labeled_dataset_4688/new_images/'
dataset2['image'] = dataset2['image'].apply(lambda x: base_path2 + x)

# combine datasets
combined_df = pd.concat([dataset1, dataset2]).reset_index(drop=True)
# reorder columns
combined_df = combined_df[['tweet_id', 'img_id', 'posted_at', 'username', 'author_location', 'text','image', 'label', 'contains_personal_info', 'country_code', 'country', 'full_name', 'author_id', 'author_name', 'media_keys']]

# fix dtypes
combined_df['author_id'] = combined_df['author_id'].astype('Int64')
combined_df['author_id'] = combined_df['author_id'].where(combined_df['author_id'].isna(), combined_df['author_id'].astype(str))
combined_df = combined_df.applymap(lambda x: np.nan if pd.isna(x) else x)
combined_df = combined_df.astype(str)
```
